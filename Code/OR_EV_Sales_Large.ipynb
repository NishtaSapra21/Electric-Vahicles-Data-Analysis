```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from data_cleaning import *
```

```python
# Read CSV file 
ev_sales_or_large = pd.read_csv("./CSVs/OREVDashboardData.csv")
ev_sales_or_large
```

```python
# Get all information about DataFrame 
checking_dataframe(ev_sales_or_large)
```

```python
# Remove Columns

clean_ev_sales_or_large = ev_sales_or_large.drop(['reg_period','ave_mo_vmt','County_GIS','SubCounty','TRACTCE10','BLOCKCE10','GEOID10',
                       'TRACTCE','BLKGRPCE','AFFGEOID_1','GEOID_1','NAME10','Senate District','House District',
                       'Congressional District','exempt_flg','Index_' ], axis=1)

clean_ev_sales_or_large
```

```python
# Clening column names 
clean_columns(clean_ev_sales_or_large)
clean_ev_sales_or_large
```

```python
# Remove whitespace from entire Dataframe
clean_ev_sales_or_large.applymap(remove_whitespace)
```

```python
# Drop Null Values
clean_ev_sales_or_large = clean_ev_sales_or_large.dropna()
checking_dataframe(clean_ev_sales_or_large)
```

```python
# Use pandas.to_datetime() to convert string to datetime format
pd.options.mode.chained_assignment = None
clean_ev_sales_or_large["Firstregistereddate"] = pd.to_datetime(clean_ev_sales_or_large["Firstregistereddate"])
clean_ev_sales_or_large["Lastregistereddate"] = pd.to_datetime(clean_ev_sales_or_large["Lastregistereddate"])
clean_ev_sales_or_large["Registrationextensiondate"] = pd.to_datetime(clean_ev_sales_or_large["Registrationextensiondate"])

clean_ev_sales_or_large
```

```python
# Check the Data Types
clean_ev_sales_or_large.dtypes
```

```python
# Rename specific column names
pd.options.mode.chained_assignment = None
clean_ev_sales_or_large.rename(columns={'Firstregistereddate':'First_Reg_Date','Lastregistereddate':'Last_Reg_Date',
                                        'Registrationextensiondate':'Reg_Extn_Date'}, inplace=True)

clean_ev_sales_or_large.head()
```

```python
# Total number of vehicle sold by type 
clean_ev_sales_or_large["Vehicle_Ty"].value_counts()
```

```python
# Total number of EV and ZEV vehicles sales in Oregon
clean_ev_sales_or_large["Zev_Type"].value_counts()
```

```python
# Top 5 EV makers in Oregon
clean_ev_sales_or_large["Make"].value_counts().head()
```

```python
# Top 5 EV model sales in Oregon
clean_ev_sales_or_large["Model"].value_counts().head()
```

```python
# Countrywise EV sales by makers
countywise_make = clean_ev_sales_or_large.groupby(['County','Make']).size()
display(countywise_make)
```

```python
# Write clean CSV file
clean_ev_sales_or_large.to_csv("./CSVs/clean_datasets/ev_sales_or_large.csv",index=False)
```
