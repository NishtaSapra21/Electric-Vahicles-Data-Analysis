```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from data_cleaning import *
```

```python
# Read CSV file 
fuel_prices = pd.read_csv("./CSVs/Fuel_Prices_Over_The_Years.csv")
fuel_prices
```

```python
# Get all information about DataFrame 
checking_dataframe(fuel_prices)
```

```python
# Clening column names 
clean_columns(fuel_prices)
```

```python
# Remove whitespace from entire Dataframe
clean_fuel_prices = fuel_prices.applymap(remove_whitespace)

clean_fuel_prices
```

```python
# Check DataFrarme
checking_dataframe(clean_fuel_prices)
```

```python
# Use pandas to_datetime() to convert string to datetime format
clean_fuel_prices["Survey_Start_Date"] = pd.to_datetime(clean_fuel_prices["Survey_Start_Date"])
clean_fuel_prices
```

```python
# Replace "na" with "NaN" in DataFrame
clean_fuel_prices = clean_fuel_prices.replace(r'^[a-zA-Z]', np.nan, regex=True)
```

```python
# Convert prices of fuels from string to float
clean_fuel_prices[clean_fuel_prices.columns[1:]] = clean_fuel_prices[clean_fuel_prices.columns[1:]].apply(lambda x: x.str.replace('$','')).astype(np.float64)
clean_fuel_prices
```

```python
# Check DataFrame
checking_dataframe(clean_fuel_prices)
```

```python
# Write clean CSV file
clean_fuel_prices.to_csv("./CSVs/clean_datasets/fuel_prices.csv", index=False)
```

```python
# Get average fuel prices yearwise
fuel_prices_yearwise = round(clean_fuel_prices.groupby(pd.Grouper(key='Survey_Start_Date', freq='Y')).mean(),2)
fuel_prices_yearwise
```

```python
min_price = []
max_price = []
percntg_change = []
fuel_name = []

columns= fuel_prices_yearwise
for col in columns:
    min_price_data = fuel_prices_yearwise[col].min()
    min_price.append(min_price_data)
    max_price_data = fuel_prices_yearwise[col].max()
    max_price.append(max_price_data)
    percntg_change_data = ((max_price_data-min_price_data)/min_price_data)*100
    percntg_change.append(round(percntg_change_data,2))
    fuel_name.append(fuel_prices_yearwise[col].name)

data = {'Fuel':fuel_name,'Min Price':min_price, 'Max Price':max_price,'% Change':percntg_change}
# Create DataFrame  
fuel_price_change = pd.DataFrame(data)  
print(fuel_price_change.to_string(index = False))
```

```python
# Plotting line chart for fuel prices through the years 

# Creating subplot/axes
fig, ax = plt.subplots()

# Setting axes/plot title
ax.set_title('Fuel Prices Yearwise')

# Setting X-axis and Y-axis labels
ax.set_xlabel('Years')
ax.set_ylabel('Fuel Prices ($)')

loc = mdates.YearLocator()
ax.xaxis.set_major_locator(loc)
fmt = mdates.DateFormatter('%Y')
ax.xaxis.set_major_formatter(fmt)

for tick in ax.get_xticklabels():
    tick.set_rotation(63)
    
for column in fuel_prices_yearwise:
    ax.plot(fuel_prices_yearwise.index,fuel_prices_yearwise[column],label=column)  
    
fig.tight_layout() #tight margins
leg = ax.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)
plt.draw() 
```
